```python
##Importing the necessary libraries

import requests
import pandas as pd
import seaborn as sns
from googleapiclient.discovery import build
from IPython.core.display import JSON
import matplotlib.pyplot as plt

##storing all your necessary credential keys

API_KEY = "AIzaSyCZXRGn38VOEZTGQw1Sve0JJmYWQFPcmwY"
Channel_ID = "UCDybamfye5An6p-j1t2YMsg"

youtube = build("youtube", "v3", developerKey = API_KEY )

##Building a function to get the necessary channels details and stats

def get_channel_stats(youtube, Channel_ID):
    request = youtube.channels().list(part= "snippet, statistics, contentDetails", id = Channel_ID)
    response = request.execute()
    data = dict(channel_name =response['items'][0]['snippet']['title'],
    date_posted = response['items'][0]['snippet']['publishedAt'],
    view_count= response['items'][0]['statistics']['viewCount'],
    subscribers = response['items'][0]['statistics']['subscriberCount'],
    video_count = response['items'][0]['statistics']['videoCount'],
    playlist_id = response['items'][0]['contentDetails']['relatedPlaylists']['uploads'])

    return data

##extracting the playlist_id
stats= get_channel_stats(youtube, Channel_ID)
playlist_id = stats['playlist_id']

##Building a function to get the video_ids from playlistItems
def get_videoids (youtube, playlist_id):
  request = youtube.playlistItems().list(part = "contentDetails", playlistId= playlist_id, maxResults = 50)
  response = request.execute()
  all_videos= []


  for i in range(len(response['items'])):
    video_ids = response['items'][i]['contentDetails']['videoId']
    all_videos.append(video_ids)
  pagetoken = response.get('nextPageToken')
  more_pages= True

  while more_pages:
      if pagetoken is None:
        more_pages = False
      else:
        request = youtube.playlistItems().list(part = "contentDetails", playlistId= playlist_id,pageToken = pagetoken, maxResults = 50)
        response = request.execute()
        for i in range(len(response['items'])):
          video_ids = response['items'][i]['contentDetails']['videoId']
          all_videos.append(video_ids)
          pagetoken = response.get('nextPageToken')
  return all_videos



##Building a function to get the videos together with the necessary details and stats
def get_videos(youtube, video_ids):
  videos = []
  for i in range(0, len(video_ids), 50):
    request = youtube.videos().list(part = "snippet,statistics, contentDetails", id = ','.join(video_ids[i:i+50]))
    response = request.execute()

    for video in range(len(response['items'])):
      video_stats = dict(title = response['items'][video]['snippet']['title'],
      publishedAt = response['items'][video]['snippet']['publishedAt'],
      views =response['items'][video]['statistics']['viewCount'],
      likes = response['items'][video]['statistics']['likeCount'],
      comment_count =response['items'][video]['statistics']['commentCount'],
      duration = response['items'][video]['contentDetails']['duration'])

      videos.append(video_stats)

  return videos
##Creating a data frame using pandas so that we can do further explorartory data analysis
video_ids = get_videoids(youtube, playlist_id)
videos = get_videos(youtube, video_ids)
df = pd.DataFrame(videos)
df.dtypes

##changing the columns to the right type
!pip install isodate
import isodate

def convert_duration_to_minutes(duration):
    # Parse the ISO 8601 duration string into a time object
    duration_obj = isodate.parse_duration(duration)

    # Convert the time object to total minutes
    total_minutes = duration_obj.total_seconds() / 60

    return total_minutes

df['duration'] = df['duration'].apply(convert_duration_to_minutes)

df['publishedAt'] = pd.to_datetime(df['publishedAt'])

df['publishedDate'] = df['publishedAt'].dt.date

df['publishedTime'] = df['publishedAt'].dt.time

df['publishedHour'] = df['publishedAt'].dt.hour

df['day_name'] = df['publishedAt'].dt.strftime('%A')

df['month_name'] = df['publishedAt'].dt.strftime('%B')

df['views'] = pd.to_numeric(df['views'], errors='coerce')

df['likes'] = pd.to_numeric(df['likes'], errors='coerce')

df['comment_count'] = pd.to_numeric(df['comment_count'], errors='coerce')



sns.barplot(x='publishedHour', y='views', data=df)


```
